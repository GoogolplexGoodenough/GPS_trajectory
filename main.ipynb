```python
from utils import read_file
import glob
import numpy as np

files = glob.glob('data/*')
data = read_file(files[1])
lat = [float(d[2]) for d in data]
lon = [float(d[3]) for d in data]
center = [np.mean(lat), np.mean(lon)]
```

```python
import folium
import pandas as pd
```

```python
center_map = folium.Map(attr='https://ditu.amap.com/abs',location=center, zoom_start=18)
```

```python
folium.PolyLine(np.column_stack((lat, lon)).tolist(),
                weight=5,       # 粗细
                opacity=0.8,    # 透明度
                color='orange').add_to(center_map)

# for idx, (la, lo) in enumerate(zip(lat, lon)):
#     folium.Marker([la,lo],popup='<b>checkpoint</b>').add_to(center_map)
    
center_map
```
